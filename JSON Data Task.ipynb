# Creating new columns with JSON Data

```python
import requests
import pandas as pd
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
```

```python
data=requests.get(url)
```

```python
data
```

```python
type(data)
```

```python
dfnew=data.json()
```

```python
type(dfnew)
```

```python
dfnew[0]['user']['login']
```

```python
idnew=[]
loginnew=[]
node_idnew=[]
avatar_urlnew=[]
urlnew=[]
followers_urlnew=[]
following_urlnew=[]
html_urlnew=[]
gists_urlnew=[]
starred_urlnew=[]
subscriptions_urlnew=[]
organizations_urlnew=[]
repos_urlnew=[]
events_urlnew=[]
received_events_urlnew=[]
typenew=[]
site_adminnew=[]
for i in range(len(dfnew)):
    loginnew.append(dfnew[i]['user']['login'])
    idnew.append(dfnew[i]['user']['id'])
    node_idnew.append(dfnew[i]['user']['node_id'])
    avatar_urlnew.append(dfnew[i]['user']['avatar_url'])
    urlnew.append(dfnew[i]['user']['url'])
    followers_urlnew.append(dfnew[i]['user']['followers_url'])
    following_urlnew.append(dfnew[i]['user']['following_url'])
    html_urlnew.append(dfnew[i]['user']['html_url'])
    gists_urlnew.append(dfnew[i]['user']['gists_url'])
    starred_urlnew.append(dfnew[i]['user']['starred_url'])
    subscriptions_urlnew.append(dfnew[i]['user']['starred_url'])
    organizations_urlnew.append(dfnew[i]['user']['starred_url'])
    repos_urlnew.append(dfnew[i]['user']['starred_url'])
    events_urlnew.append(dfnew[i]['user']['events_url'])
    received_events_urlnew.append(dfnew[i]['user']['received_events_url'])
    typenew.append(dfnew[i]['user']['type'])
    site_adminnew.append(dfnew[i]['user']['site_admin'])
```

```python
data1=pd.DataFrame(dfnew)
```

```python
data1['login_new']=pd.DataFrame(loginnew)
data1['id_new']=pd.DataFrame(idnew)
data1['node_id_new']=pd.DataFrame(node_idnew)
data1['avatar_url_new']=pd.DataFrame(avatar_urlnew)
data1['url_new']=pd.DataFrame(urlnew)
data1['html_url_new']=pd.DataFrame(html_urlnew)
data1['gists_url_new']=pd.DataFrame(gists_urlnew)
data1['starred_url_new']=pd.DataFrame(starred_urlnew)
data1['subscriptions_url_new']=pd.DataFrame(subscriptions_urlnew)
data1['organizations_url_new']=pd.DataFrame(organizations_urlnew)
data1['repos_url_new']=pd.DataFrame(repos_urlnew)
data1['events_url_new']=pd.DataFrame(events_urlnew)
data1['received_events_url_new']=pd.DataFrame(received_events_urlnew)
data1['type_new']=pd.DataFrame(typenew)
data1['site_admin_new']=pd.DataFrame(site_adminnew)
```
